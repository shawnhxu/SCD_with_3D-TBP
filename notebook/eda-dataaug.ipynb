## Data Augmentation step

Due to the massive imbalance of positive and negative samples in the dataset, there is a clear problem already before even selecting a model to train. Due to this, there needs to be steps to either address the lack of positive images through data augmentation/image generation based on currently supplied positive images or other novel algorithms. 

Options:
- SMOTE to generate more of the minority class
- https://link.springer.com/article/10.1007/s12559-023-10187-8 (wind turbine failure detection -- novel ensemble-learning based CNN for imbalanced data)
- https://ieeexplore.ieee.org/abstract/document/9681083 (An Improved Extreme Learning Machine for Imbalanced Data Classification)
- https://github.com/ufoym/imbalanced-dataset-sampler (Imbalanced Dataset sampler)
